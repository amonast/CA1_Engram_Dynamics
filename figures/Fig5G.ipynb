```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
plt.style.use('/Users/amonast/Documents/GitHub/CA1_Engram_Dynamics/figures/paper_style.mplstyle')
import pingouin as pg
```

```python
df = pd.read_csv('/Users/amonast/Documents/GitHub/Amy_Reviews/data/riem_dist_tfc_recall1.csv')
```

```python
palette = {
    'EE': '#F37243',     
    'NN': '#00ABC8'}
gray = '#424949'
```

```python
df.head()
```

```python
fig,ax=plt.subplots(figsize=(2.4,2.5))
sb.boxplot(data = df,x='pair_type',y='distance',palette=palette,ax=ax,gap=0.2,
            width=.7, linewidth=1.5, showfliers=False)
sb.stripplot(data = df,x='pair_type',y='distance',palette=palette,ax=ax,
             edgecolor=gray, linewidth=1.5, dodge=False, jitter=0, alpha=.7, size=5, legend=False)
sb.lineplot(data = df,x='pair_type',y='distance',color=gray,ax=ax,units='animal',estimator=None,
             linewidth=1, alpha=.7, legend=False)
ax.set_ylabel(f'Pre-CS+ $\\rightarrow$ CS+\nRiemannian Distance')
ax.set_xlabel('Correlation')
ax.hlines(.26,0,1,colors='k',linewidth=1)
ax.text(.5,.26,'***',ha='center',size=10)

sb.despine()
plt.tight_layout()
plt.savefig('Fig5G_Recall1_RiemDist.svg',transparent=True)
```
