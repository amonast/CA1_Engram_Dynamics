```python
def get_decile_df(ani,fov,session,file_key,base_dir):
    def calculate_deciles_ani(ani_df):
        animal = ani_df['Animal'].values[0]
        ani_df['Decile']=pd.qcut(ani_df['Event Rate'], 10,labels=False)
        # ani_df['Decile']=ani_df['Decile']+1
        decile_ptagged = [ani_df.loc[(ani_df['Decile']==d)&(ani_df['Population']=='Tagged')].shape[0] 
                    / ani_df.loc[(ani_df['Decile']==d)].shape[0] for d in np.unique(ani_df['Decile'])]
        decile_means = [ani_df['Event Rate'].loc[ani_df['Decile']==d].mean() for d in np.unique(ani_df['Decile'])]
        df = pd.DataFrame()
        df['Animal']=[animal]*10
        df['Decile Mean Rates']=decile_means
        df['P(Tagged)']=decile_ptagged
        return df,ani_df

    ani='992N'
    df = pd.read_csv('/Users/amonast/Desktop/Spont2P/Analysis/rates/rates_all_mice_df_weight_5std_filt.csv')
    info = pd.read_csv('/Users/amonast/Desktop/Spont2P/Data_info.csv')
    df = df.drop(['Unnamed: 0'],axis=1)
    ani_df = df.loc[(df['Animal']==ani)&(df['Session']=='Baseline')].reset_index(drop=True)
    _,cells_df= calculate_deciles_ani(ani_df)
    decile_df = cells_df.loc[(cells_df.FOV=='FOV2')&(cells_df.Baseline!=-1)]
    return decile_df
```

```python
file_key = '/Volumes/AM_SSD1/Spont2P/Data_info.csv'
base_dir = '/Volumes/AM_SSD1/Spont2P'
ani,fov='992N','FOV2'
_, mean_list, _, cnmf_list = get_fov_data(ani,fov,file_key,base_dir)
image = mean_list[0]
cnm = cnmf_list[0]
decile_df = get_decile_df(ani,fov,'Baseline',file_key,base_dir)
colors = np.arange(0,1,10)
```

```python
def get_sorted_masks(cnm,decile_df,population=None,colors=None):
    masks_list= []
    if population is None:
        for d in range(1,10):
            inds = decile_df.Baseline.loc[decile_df.Decile==d].values
            masks_decile = get_mask_rois(cnm,idx=inds)
            masks_decile[masks_decile==1]=colors[d]
            masks_list.append(masks_decile)
        masks = np.concatenate(masks_list,axis=0).max(axis=0)
        masks[masks == 0] = np.nan
        return masks
    else:
        for d in range(1,10):
            inds = decile_df.Baseline.loc[(decile_df.Decile==d)&(decile_df.Population==population)].values
            masks_decile = get_mask_rois(cnm,idx=inds)
            masks_decile[masks_decile==1]=colors[d]
            masks_list.append(masks_decile)
        masks = np.concatenate(masks_list,axis=0).max(axis=0)
        masks[masks == 0] = np.nan
        return masks
```

```python
tag = get_sorted_masks(cnm,decile_df,colors= np.linspace(0,1,11),population='Tagged')
non = get_sorted_masks(cnm,decile_df,colors= np.linspace(0,1,11),population='Non-Tagged')
```
