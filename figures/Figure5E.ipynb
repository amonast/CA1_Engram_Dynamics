```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import pingouin as pg
plt.style.use('/Users/amonast/Documents/GitHub/CA1_Engram_Dynamics/figures/paper_style.mplstyle')
```

```python
ani_trials_z = pd.read_csv("/Users/amonast/Desktop/Tone2P/Analysis/rates/rate_trials_Recall1_animals_norm2baseline.csv")
```

```python

hue_order = [False,True]
palette=['#00ABC8','#F37243']

##################### Mean across trials
fig,ax=plt.subplots(ncols=1,nrows=1,figsize=(4,2.8))
sb.pointplot(data=ani_trials_z,y='trial_rate_z',x='trial',hue='is_engram',errorbar='se',
                capsize=0.1,
                markersize=6,
                palette=palette,
                markers='o',
                linewidth=2,
                dodge=0,
                legend=True,
                err_kws=dict(lw=1.))
plt.legend(frameon=False,bbox_to_anchor=(.9,1))
plt.xlabel('Trial')
plt.ylabel('Event Rate (z-score)')
sb.despine()
plt.tight_layout()
plt.savefig("Fig5E_zscore_across_trials.svg" ,transparent=True)      
#%%
```

```python
pg.rm_anova(data=ani_trials_z,within=['trial','is_engram'],dv='trial_rate_z',subject='animal')
```
