```python
import matplotlib.pyplot as plt
import os 

def latex2image(latex_expression, image_name, image_size_in=(3, 0.2), fontsize=16, dpi=200, format='svg', save=False, savepath=None):
    """
    A simple function to generate an image from a LaTeX language string.

    Parameters
    ----------
    latex_expression : str
        Equation in LaTeX markup language.
    image_name : str or path-like
        Full path or filename including filetype.
    Accepted filetypes include: png, pdf, ps, eps and svg.
    image_size_in : tuple of float, optional
        Image size. Tuple which elements, in inches, are: (width_in, vertical_in).
    fontsize : float or str, optional
        Font size, that can be expressed as float or
        {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}.

    Returns
    -------
    fig : object
        Matplotlib figure object from the class: matplotlib.figure.Figure.
    """
    if isinstance(latex_expression, str):
        fig, ax = plt.subplots(figsize=image_size_in, dpi=dpi)
        ax.axis('off')  # Turn off the axes
        fig.text(
            x=0.5,
            y=0.5,
            s=latex_expression,
            horizontalalignment="center",
            verticalalignment="center",
            fontsize=fontsize
        )
    elif isinstance(latex_expression, list):
        fig, ax = plt.subplots(figsize=(image_size_in[0], len(latex_expression) * image_size_in[1]), dpi=dpi)
        ax.axis('off')  # Turn off the axes
        for i, exp in enumerate(latex_expression):
            fig.text(
                x=0.5,
                y=0.5 - i * 0.1,  # Adjust y-position for each line
                s=exp,
                horizontalalignment="center",
                verticalalignment="center",
                fontsize=fontsize
            )
    plt.tight_layout()
    if save:
        if savepath is None:
            plt.savefig(image_name + '.' + format, bbox_inches='tight', pad_inches=0,transparent=True))
        else:
            plt.savefig(os.path.join(savepath, image_name + '.' + format), bbox_inches='tight', pad_inches=0,transparent=True))
    plt.show()
    return fig
```

```python
eq1 =r"$p_{i,j} = \beta_{1i} \; \text{EventRate}_{i,j \;REST} +  \beta_{0\;i}$"
eq2 = r"$p_{i,j} = \beta_{1i} \; \text{EventRate}_{i,j\;RUN } +  \beta_{0\;i}$"
eq3 = r"$ p_{i,j} = \beta_{1i\;RUN}  \; \text{EventRate}_{i,j\;REST } +$"
eq4=r"$\beta_{1i\;RUN}  \text{EventRate}_{i,j\;RUN } + \beta_{0\;i}$"
```

```python
latex2image(eq1,'rest_eq',save=True,format='svg')
latex2image(eq2,'run_eq',save=True,format='svg')
latex2image(eq3,'combined_eq',save=True,format='svg')
latex2image(eq4,'combined_eq2',save=True,format='svg')
```
