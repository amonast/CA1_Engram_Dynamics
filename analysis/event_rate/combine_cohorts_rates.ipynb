```python
##### rest_rates
rest_csvs=['rates_rest_mice_df_weighted_5std_repeat.csv',
           'rates_lou_rest_mice_df_weighted_5std_repeat.csv']

df_Rest = pd.DataFrame()
for f in rest_csvs:
    file=os.path.join(data_dir,f)
    df = pd.read_csv(file)
    df_Rest = pd.concat([df_Rest,df],ignore_index=True)
#df_Rest = df_Rest[df_Rest['Event Rate'] != 0]

##### running rates
run_csvs=['rates_run_mice_df_weighted_5std_repeat.csv',
'rates_lou_run_mice_df_weighted_5std_repeat.csv']
df_Run = pd.DataFrame()
for f in run_csvs:
    file=os.path.join(data_dir,f)
    df = pd.read_csv(file)
    df_Run = pd.concat([df_Run,df],ignore_index=True)
#df_Run = df_Run[df_Run['Event Rate'] != 0]

 ##### all time
tot_csvs=['rates_all_mice_df_weight_5std_filt.csv',
          'rates_lou_mice_df_weight_5std_filt.csv']
df_Tot = pd.DataFrame()
for f in tot_csvs:
    file=os.path.join(data_dir,f)
    df = pd.read_csv(file)
    df_Tot = pd.concat([df_Tot,df],ignore_index=True)
#df_Tot= df_Tot[df_Tot['Event Rate'] != 0]
```

```python
df_Tot.to_csv('/Users/amonast/Documents/GitHub/Amy_Reviews/data/rates_total_weighted_5std_all.csv')
df_Rest.to_csv('/Users/amonast/Documents/GitHub/Amy_Reviews/data/rates_rest_weighted_5std_all.csv')
df_Run.to_csv('/Users/amonast/Documents/GitHub/Amy_Reviews/data/rates_run_weighted_5std_all.csv')
```
